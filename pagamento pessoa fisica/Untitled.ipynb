```python
def galera(quem):
    
    if type(quem) == str:
        
        quem = df[df.NOME == quem]
    
    totais_por_mes = quem.groupby(['NOME', quem.DATA.dt.year, quem.DATA.dt.month])['VALOR (R$)'].sum().unstack()
    totais_por_mes['total'] = totais_por_mes.sum(axis=1)
    
    return totais_por_mes
```

```python
projeto = df[df.PROJETO == '1520 - CT - CAGECE/FCPC/PORTO DO PECEM']

galera(projeto)
```

```python
barone = df[df.NOME.str.contains("BARONE")]

galera(barone)
```

```python
serv_ufc = pd.read_csv("../../dataset-remuneracao-federal/NOME.csv")
```

```python
df_ufc = df[df.NOME.isin(serv_ufc.NOME)]
```

```python
df_ufc['TIPO DE PAGAMENTO'].unique()
```

```python
rpa = df_ufc[df_ufc['TIPO DE PAGAMENTO']\
                       .isin(['RPA', 'SERVIÇO PRESTADO - RPA', 'SERVIÇOS DE TERCEIROS - PESSOA FÍSICA'])]
top_3_ufc_rpa = rpa\
.groupby('NOME')['VALOR (R$)']\
.sum().nlargest(3).index

galera(rpa[rpa.NOME.isin(top_3_ufc_rpa)])
```
